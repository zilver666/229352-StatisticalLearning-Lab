Ada

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
import matplotlib.pyplot as plt

param = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1.0]
}

ada = AdaBoostClassifier(random_state=42)
gs = GridSearchCV(ada, param, cv=5)
gs.fit(X_train, y_train)

print(gs.best_score_)
print(gs.score(X_test, y_test))

plt.bar(range(len(gs.best_estimator_.feature_importances_)), gs.best_estimator_.feature_importances_)
plt.show()
```

XG

```python
from xgboost import XGBClassifier, plot_importance

param = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.01, 0.1]
}

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', verbosity=0)
gs = GridSearchCV(xgb, param, cv=5)
gs.fit(X_train, y_train)

print(gs.best_score_)
print(gs.score(X_test, y_test))

plot_importance(gs.best_estimator_)
plt.show()
```

Light

```python
from lightgbm import LGBMClassifier, plot_importance

param = {
    'n_estimators': [100, 200],
    'num_leaves': [20, 31],
    'learning_rate': [0.05, 0.1]
}

lgbm = LGBMClassifier(verbose=-1)
gs = GridSearchCV(lgbm, param, cv=5)
gs.fit(X_train, y_train)

print(gs.best_score_)
print(gs.score(X_test, y_test))

plot_importance(gs.best_estimator_)
plt.show()
```
